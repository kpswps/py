###### 5. Policz średnią dla każdej kolumny z zarobkami. Co trzeba zrobić, żeby móc wykonać tę operację? Stwórz listę ze średnimi.

```python
dollar_cols = ['bgn_p50','mid_p50',
               'mid_p10','mid_p25',
               'mid_p75','mid_p90']
```

```python
for each_col in dollar_cols:
    dfs[0][each_col] = dfs[0][each_col].str.replace("$", "")
    dfs[0][each_col] = dfs[0][each_col].str.replace(",", "")
    dfs[0][each_col] = pd.to_numeric(dfs[0][each_col])
```
