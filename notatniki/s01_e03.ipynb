#### Czego listy nie zrobią?

```python
import numpy as np
import time
```

```python
#wszyscy losujemy takie same dane
seed = np.random.RandomState(100)
wzrost_lista = list(seed.normal(loc=1.70,scale=.15,size=100000).round(2))
seed2 = np.random.RandomState(100)
waga_lista = list(seed2.normal(loc=80,scale=10,size=100000).round(2))
```

```python
#Wzrost miał być zaokrąglony do 2 miejsc po przecinku, niestety w liście róznie to działa.
wzrost_lista
```

```python
#Czy da się policzyć cokolwiek z całej listy (bez brania pojedynczych elementów)? Nie.
bmi_list = waga_lista / wzrost_lista**2
```

```python
#Jak policzyć BMI dla wagi i wzrostu podanych w liście?

start_time = time.clock()
bmi_lista = list() #Stwórz prostą listę
for index, each in enumerate(range(len(wzrost_lista))): #Dla każdego elementu ze zbioru od 0 do 100000 (ale bez niego....)
    result = waga_lista[index] / wzrost_lista[index]**2 #Policz mi dla danego elementu BMI
    bmi_lista.append(result) # Dodaj wynik do pustej listy
print(time.clock() - start_time, "seconds") #Tyle czasu to zajęło dla 100000 elementów. Tragedii nie ma!
```

#### Rozwiązanie - NumPy - Wektory, macierze.

```python
import numpy as np
```

```python
#Jak stworzyć dowolny wektor/macierz? Do poniższej funkcji wpisujemy listę z wartościami.
np.array([3,5,9])
```

```python
#wszyscy losujemy takie same dane
seed = np.random.RandomState(100)
wzrost_array = seed.normal(loc=1.70,scale=.15,size=100000).round(2)
seed2 = np.random.RandomState(100)
seed3 = np.random.RandomState(50)
waga_array = seed2.normal(loc=80,scale=10,size=100000).round(2) + seed3.normal(loc=9,scale=9,size=100000).round(2)
```

```python
wzrost_array
```

```python
start_time = time.clock()
bmi_array = waga_array / wzrost_array**2 #jak liczyć BMI przy pomocy NumPy? O tak.
print(time.clock() - start_time, "seconds")
```

#### Podzbiory

```python
#niestety z listy nie wybierzemy też elementów na podstawie jakiegoś warunku!
wzrost_lista > 1.60
```

```python
#tak też nie pójdzie..
wzrost_lista[wzrost_lista > 1.60]
```

```python
#Ok, tak idzie - dostajemy wektor wartości logicznych - True: ktoś spełnia nasze warunki; False - nie spełnia.
wzrost_array > 1.60
```

```python
#pokaż mi wzrost tych ludzi, którzy mierzą więcej niż 1.60m!
wzrost_array[wzrost_array > 1.60]
```

```python
wysokie_ludzie = wzrost_array[wzrost_array > 1.60]
```

```python
#tyle osób spełniło nasz warunek wzrostu!
len(wysokie_ludzie)
```

```python
#Wzrost tylko dla ciężkich? Żaden problem - pokaż mi wzrost dla tych, którzy ważą więcej niż 100kg!
wzrost_array[waga_array > 100]
```

```python
#Zapisanie wektora wartości logicznych do zmiennej? Nie ma sprawy!
jak_oni_zyja = bmi_array >= 50
```

```python
jak_oni_zyja
```

```python
#Nie musimy wpisywać warunku, możemy się odnieść do naszej zmiennej jak_oni_zyja!
bmi_array[jak_oni_zyja]
```

```python
wzrost_array[jak_oni_zyja]
```

```python
#dwa warunki? żaden problem!
#pokaż mi bmi tych ludzi, którzy ważą więcej niż 60kg, ale mniej niż 80kg!

pomiedzy = bmi_array[(waga_array > 60) & (waga_array < 80)]
```

```python
pomiedzy
```

##### Bonus na koniec - przykładowy wykres!

```python
import matplotlib.pyplot as plt
plt.plot(wzrost_array[bmi_array >= 50], waga_array[bmi_array >= 50], 'x',color='green',label='BMI >= 50')
plt.plot(wzrost_array[(bmi_array <= 25) & (bmi_array >= 24)], waga_array[(bmi_array <= 25) & (bmi_array >= 24)], 'x',color='red', label='25≥BMI≥24')
plt.plot(wzrost_array[bmi_array < 20], waga_array[bmi_array < 20], 'x',color='blue', label='BMI < 20')
plt.legend()
plt.show()
```
